### Q1.8 Magnitude Calculations

```python
#Sirius data for testing: m = -1.46, M = 1.45

#1. Get interactive input for the magnitudes
m = float(input("Enter the apparent magnitude (m): "))
M = float(input("Enter the absolute magnitude (M): "))

#2. Calculate the distance in parsecs using said formula: d = 10**((m-M)/5 + 1)
d_parsec = 10.0((m - M) / 5.0 + 1.0)

#3. Convert parsecs to lightyears (1 pc = 3.26.164 lightyears)
d_ly = d_parsec * 3.26164

#4. Print the result using an f-string
print(f"The resulting distance is {d_ly:.2f} lightyears.")
```
