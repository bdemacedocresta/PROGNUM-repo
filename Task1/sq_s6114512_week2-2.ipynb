```python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

set1 = set(night1)
set2 = set(night2)

intersection_list = list(set1.intersection(set2))
print(f"Intersection list: {intersection_list}")

x = 5 in set1
print(x)

x = 5 in intersection_list
print(x)
```
